```python
filename = 'encoded_text'
file = open(filename, encoding='cp1251')
```

```python
file
```

```python
file.readlines()
```

```python
lines = file.readlines()
```

```python
for line in lines:
    words = 0
    line_num = 1
    for word in line.split(' '):
        if word[0].upper() == word[0]:
            print(word[0])
            words += 1
    line_num += 1
    print('Строка ' + str(line_num) + ' имеет в себе ' + str(words) + ' слова с большой буквой')
```

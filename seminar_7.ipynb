#### Записать в файл реплики персонажа, чье имя вобьем в инпут

```python
chname = input().lower()
with open('../../all_subtitles.txt', encoding='utf-8') as subs:
    with open('chlines.txt', 'w', encoding='utf-8') as chlines:
        for line in subs:
            if line[0:len(chname)] == chname:
                chlines.write(line)
```

```python
file = open('chlines.txt', 'r', encoding='utf-8')
stud = file.readlines()
len(stud)
```

#### Кто был последним, кто говорил реплику в сериале, потом выводим принтом первую реплику

```python
with open('../../all_subtitles.txt', encoding='utf-8') as subs:
    name = ''
    for line in subs:
        if len(line.split(':')) > 1:
            name = line.split(':')[0]
            
with open('../../all_subtitles.txt', encoding='utf-8') as subs:
    for line in subs:
        if name+': ' in line:
            print(line)
            break
```
